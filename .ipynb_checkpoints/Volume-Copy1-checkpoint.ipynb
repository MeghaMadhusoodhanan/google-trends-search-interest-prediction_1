```python
# Core libraries
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Modeling
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# Evaluation
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.inspection import permutation_importance
```

```python
# Load dataset
df = pd.read_csv(r"C:\Users\ACER\OneDrive\Desktop\us trend\trending_searches_in_us.csv")

df.head()
```

```python
df.describe()
```

```python
# Convert date columns safely
df['start_date'] = pd.to_datetime(df['start_date'], errors='coerce')
df['end_date'] = pd.to_datetime(df['end_date'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
# Drop rows with critical missing values
df = df.dropna(subset=['start_date', 'increase_percentage', 'search_volume'])
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[column] >= lower) & (data[column] <= upper)]

df = remove_outliers_iqr(df, 'increase_percentage')
df = remove_outliers_iqr(df, 'search_volume')
```

```python
# Extract time-based features
df['year'] = df['start_date'].dt.year
df['month'] = df['start_date'].dt.month
df['day'] = df['start_date'].dt.day
df['dayofweek'] = df['start_date'].dt.dayofweek
```

```python
X = df[['increase_percentage', 'year', 'month', 'day', 'dayofweek']]
y = df['search_volume']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)
```

```python
lr_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LinearRegression())
])

lr_pipeline.fit(X_train, y_train)
```

```python
svr_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', SVR())
])

svr_params = {
    'model__C': [0.1, 1, 10],
    'model__epsilon': [0.01, 0.1],
    'model__kernel': ['rbf']
}

svr_grid = GridSearchCV(
    svr_pipeline,
    svr_params,
    cv=3,
    scoring='r2',
    n_jobs=-1
)

svr_grid.fit(X_train, y_train)
```

```python
dt_params = {
    'max_depth': [5, 10, 20, None],
    'min_samples_split': [2, 5, 10]
}

dt_grid = GridSearchCV(
    DecisionTreeRegressor(random_state=42),
    dt_params,
    cv=3,
    scoring='r2'
)

dt_grid.fit(X_train, y_train)
```

```python
rf_params = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None]
}

rf_grid = GridSearchCV(
    RandomForestRegressor(random_state=42),
    rf_params,
    cv=3,
    scoring='r2',
    n_jobs=-1
)

rf_grid.fit(X_train, y_train)
```

```python
def evaluate(model, X_test, y_test, name):
    preds = model.predict(X_test)
    print(f"\n{name}")
    print("R:", r2_score(y_test, preds))
    print("RMSE:", np.sqrt(mean_squared_error(y_test, preds)))
    print("MAE:", mean_absolute_error(y_test, preds))
```

```python
evaluate(lr_pipeline, X_test, y_test, "Linear Regression")
evaluate(svr_grid.best_estimator_, X_test, y_test, "SVR")
evaluate(dt_grid.best_estimator_, X_test, y_test, "Decision Tree")
evaluate(rf_grid.best_estimator_, X_test, y_test, "Random Forest")
```

```python
best_model = rf_grid.best_estimator_

perm = permutation_importance(
    best_model, X_test, y_test,
    n_repeats=10,
    random_state=42
)

importance_df = pd.DataFrame({
    'feature': X.columns,
    'importance': perm.importances_mean
}).sort_values(by='importance', ascending=False)

importance_df
```

```python
summary = pd.DataFrame({
    'Model': ['Linear Regression', 'SVR', 'Decision Tree', 'Random Forest'],
    'Best R2': [
        r2_score(y_test, lr_pipeline.predict(X_test)),
        r2_score(y_test, svr_grid.best_estimator_.predict(X_test)),
        r2_score(y_test, dt_grid.best_estimator_.predict(X_test)),
        r2_score(y_test, rf_grid.best_estimator_.predict(X_test))
    ]
})

summary.sort_values(by='Best R2', ascending=False)
```
